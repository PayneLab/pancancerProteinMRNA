```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy 
```

```python
df = pd.read_csv('data/annotated_combinded_GO_pathway_enrichment_df.csv')
df = df.sort_values('Cancer Hallmark')
df['-log10(p_value)'] = - scipy.log10(df['p_value'])
```

```python
plt.figure(figsize=(9,13))
g = sns.barplot(data = df, x = '-log10(p_value)', y = 'name', hue = 'Cancer Hallmark')
g.legend(bbox_to_anchor=(1.01, .5), ncol=1, title = 'Cancer Hallmark')
plt.ylabel('GO:BP Pathway')
#plt.tight_layout()
plt.savefig('Figures/Figure 5) Cancer Hallmarks Pathways.png', bbox_inches='tight')
```
