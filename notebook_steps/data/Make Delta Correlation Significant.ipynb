```python
import pandas as pd
```

```python
delta_corr_df = pd.read_csv("delta_correlation_df.csv")
```

```python
delta_corr_df
```

```python
significant = []
for index, row in delta_corr_df.iterrows():
    if row["FDR"] < 0.05 and abs(row["Delta_Correlation"]) > 0.2:
        significant.append(True)
    else:
        significant.append(False)
        
delta_corr_df["Significant"] =  significant
delta_corr_df
```

```python
delta_corr_df.to_csv('delta_correlation_df_with_significance.csv', index = False)
```
