```python
import pandas as pd
import glob
```

```python
files = glob.glob('Scripts_to_make_regression_dataframes/*regressions.csv')
files
```

```python
dfs = []
for file in files:
    df = pd.read_csv(file)
    dfs.append(df)
regression_df = pd.concat(dfs)
regression_df = regression_df[regression_df.Cancer != 'PDAC']
regression_df.to_csv('regression_df.csv', index = False)
regression_df
```
