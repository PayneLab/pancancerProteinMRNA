```python
import pandas as pd
from gprofiler import GProfiler
```

```python
delta_correlation_df = pd.read_csv('delta_correlation_df.csv')
delta_correlation_df
```

```python
background_genes = list(pd.unique(delta_correlation_df.Gene))
```

```python
gp = GProfiler(return_dataframe=True)
```

```python
pathway_enrichments = []
for cancer in pd.unique(delta_correlation_df.Cancer):
    cancer_df = delta_correlation_df[delta_correlation_df.Cancer == cancer]
    sig_cancer_df = cancer_df[cancer_df.FDR < 0.05]
    sig_cancer_df['Delta_Correlation'] = abs(sig_cancer_df.Delta_Correlation)
    sig_cancer_df = sig_cancer_df.sort_values('Delta_Correlation', ascending = False)
    sig_genes = list(sig_cancer_df.Gene)
    pathway_enrichment = gp.profile(organism='hsapiens', query = sig_genes, no_iea=True,
                       ordered=True, no_evidences=False, background= background_genes, sources = ['GO:BP'])
    pathway_enrichment['Cancer'] = [cancer] * len(pathway_enrichment)
    pathway_enrichments.append(pathway_enrichment)
pathway_enrichments = pd.concat(pathway_enrichments)
pathway_enrichments.to_csv('go_bp_pathway_enrichment_df.csv', index = False)
pathway_enrichments
```
