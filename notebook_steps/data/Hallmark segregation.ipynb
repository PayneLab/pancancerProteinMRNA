# KEGG

```python
kegg_profiles = []
for cancer in pd.unique(delta_correlation_df.Cancer):
    
    cancer_df = delta_correlation_df[delta_correlation_df.Cancer == cancer]
    background_genes  = list(pd.unique(cancer_df.Gene))
    cancer_df = cancer_df[cancer_df.FDR < 0.05]
    queryS = list(pd.unique(cancer_df.Gene))
    cancer_profile = gp.profile(organism='hsapiens', query = queryS, no_iea=True, sources = ["KEGG"],
                       ordered=True, no_evidences=False, background= background_genes)
    kegg_profiles.append(cancer_profile)
```

```python
df_dict = {}
prohibited_pathways = ["Epstein-Barr virus infection",
                        "Vibrio cholerae infection",
                        "Human immunodeficiency virus 1 infection",
                        "Viral myocarditis",
                        "Arrhythmogenic right ventricular cardiomyopathy",
                        "Chronic myeloid leukemia",
                          "Rheumatoid arthritis",
                      "Epithelial cell signaling in Helicobacter pylori infection",
                       "Synaptic vesicle cycle",
                      "Endocrine and other factor-regulated calcium reabsorption",
                      "Bacterial invasion of epithelial cells",
                      "Measles"]
for i in canEnum:
 
    description = (kegg_profiles[i.value]["description"])
    term = (kegg_profiles[i.value]["term_size"])
    intersection = (kegg_profiles[i.value]["intersection_size"])
    description_with_term_query = []
    for index, value in enumerate(description):
        if (value in prohibited_pathways ):
            continue
        if (term[index] > 200):
            continue
        if ((intersection[index])*100/(term[index]) >= 15):
            description_with_term_query.append(value)#  + " (" + str(intersection[index])+ "/" + str(term[index]) + ")")
    cancer_name = i.name# + " (Intersecion/Term)"
    df_dict[cancer_name]=description_with_term_query
    maxlen = len(max(df_dict.values(), key=len))
    
    for i in df_dict:
        df_dict[i] = df_dict[i] + [math.nan] * (maxlen - len(df_dict[i])) 
    
    
    
```

```python
endo = kegg_profiles[canEnum.EN.value][['description','p_value']]
luad = kegg_profiles[canEnum.LUAD.value][['description','p_value']]
lscc = kegg_profiles[canEnum.LSCC.value][['description','p_value']]
hnscc = kegg_profiles[canEnum.HNSCC.value][['description','p_value']]
ccrcc = kegg_profiles[canEnum.CCRCC.value][['description','p_value']]
```

```python
luad['Cancer'] = ['LUAD' for i in range(len(luad.index))]
endo['Cancer'] = ['Endometrial' for i in range(len(endo.index))]
lscc['Cancer'] = ['LSCC' for i in range(len(lscc.index))]
hnscc['Cancer'] = ['HNSCC' for i in range(len(hnscc.index))]
ccrcc['Cancer'] = ['CCRCC' for i in range(len(ccrcc.index))]
```

```python
# grouper = slim_peaks.groupby('Base Sequence')['Peak RT Apex']
# grouper
# res = pd.concat((pd.Series(vals.values, name=b) for b, vals in grouper), axis=1)

kegg_df = pd.concat([endo,lscc,luad,ccrcc,hnscc])
kegg_df.p_value=-np.log(kegg_df.p_value)
kegg_df = kegg_df[kegg_df.description.isin(['Fatty acid degradation', 'Valine, leucine and isoleucine degradation', 
                                  'Fatty acid metabolism', 'p53 signaling pathway', 
                                  'Propanoate metabolism', 'beta-Alanine metabolism', 
                                  'Fatty acid biosynthesis', 'Glycolysis / Gluconeogenesis',
                                  'Butanoate metabolism', 'Glycerolipid metabolism'])]
```

```python
alt.Chart(kegg_df).mark_bar().encode(
    x = alt.X(
        'p_value',
        title = 'P value(Log)'
    ),
    y = alt.Y(
        'cancer',
        title=None,
        axis=alt.Axis(
                labels=False,
                ticks=False,
            )
    ),
    color = 'cancer',
    row=alt.Row(
        'description:O',
        sort = kegg_df['description'].values,
        spacing=0,
        header=alt.Header(
                title=None,
                labelAngle=0,
                labelAnchor="middle",
                labelOrient="left",
                labelAlign="left",
            )
#     header=alt.Header(labelAngle=0)
    )
)
```
