```python
import pandas as pd
import glob
import statsmodels.stats.multitest as ssm
```

```python
files = glob.glob('Scripts_to_make_transmutation_effects_dataframes/*effects.csv')
files
```

```python
dfs = []
for file in files:
    df = pd.read_csv(file)
    dfs.append(df)
transmutation_df = pd.concat(dfs)
dfs = []
for cancer in pd.unique(transmutation_df.Cancer):
    df = transmutation_df[transmutation_df.Cancer == cancer]
    df['FDR'] = ssm.fdrcorrection(df['P_Value'])[1]
    dfs.append(df)
transmutation_df = pd.concat(dfs)
transmutation_df = transmutation_df[transmutation_df.Cancer != 'PDAC']
transmutation_df.to_csv('transmutation_df.csv', index = False)
transmutation_df
```
