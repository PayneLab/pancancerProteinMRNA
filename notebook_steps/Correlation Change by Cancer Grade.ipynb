```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/Cancer_grades_correlations.csv')
df = df[df.Grade != 'Other']
```

```python
df
df.Cancer.unique()
```

```python
sns.set_context(rc={"font.size":16,"axes.titlesize":16,"axes.labelsize":16, 'xtick.labelsize': 12, 'ytick.labelsize': 12}) 
```

```python
g = sns.FacetGrid(data = df, col = 'Cancer', hue = 'Grade', col_wrap = 3, legend_out = True, 
                  palette={"Normal" : "royalblue","Grade I" : "red", "Grade II" : "darkred",
                           "Grade III" : "purple","Grade IV" : "black"})

g.map_dataframe(sns.kdeplot, x = 'Spearman_Rho')
g.add_legend()#(bbox_to_anchor=(.75, .25))
g.tight_layout()

plt.savefig('Figures/Figure 1) Correlation changes by cancer stage')
plt.show() 
```
