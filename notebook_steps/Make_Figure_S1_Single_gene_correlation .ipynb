# Finding mRNA versus protein correlation for a single gene

```python
import cptac
import scipy
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
import pandas as pd
import pcprutils as ut
import cptac.pancan as pc
```

Import CPTAC LUAD dataset

```python
luad = pc.PancanLuad()
```

```python
def annotate(data, **kws): #Function to annotate the correlation of each plot in the plot
    r, p = scipy.stats.pearsonr(data['Proteomics'], data['Transcriptomics'])
    ax = plt.gca()
    ax.text(.05, .9, 'R2= {:.3f}'.format(r**2),
            transform=ax.transAxes)
#     ax.set(xlabel='common xlabel', ylabel='common ylabel')
```

```python
sns.set_context(rc={"font.size":16,"axes.titlesize":16,"axes.labelsize":16, 'xtick.labelsize': 12, 'ytick.labelsize': 12}) 
```

Plot proteomic and transcriptomic scatter plots

```python
genes = ['TP53','FAT1','KANSL1','MAP10']
df = ut.get_prot_trans_df(luad)
df = df[df['Tissue'] == 'Tumor']
df = df[df.Gene.isin(genes)]
plot = sns.lmplot(y="Proteomics", x="Transcriptomics", sharex= False, sharey = False, col="Gene", data=df, col_wrap=2,line_kws={'color': 'black'}, scatter_kws={'color': '#fe831b'});
axes = plot.axes
plot.set_axis_labels("RNA Abundance", "Protein Abundance")
plot.tight_layout()
plot.map_dataframe(annotate)
plt.savefig('Figures/Figure_S1_Single_gene_correlations', dpi=600)
plt.show()

```
