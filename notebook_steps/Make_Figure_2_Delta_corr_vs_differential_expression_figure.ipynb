## Comparison of Delta Correlation and Differential Expression
We used CPTAC proteomic and transcriptomic data to make Delta Correlation
Code for Delta Correlation: 
We compare the Delta Correlation to CPTAC proteomic and transcriptomic differential expression
Code for Differential Expression:

```python
import cptac
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats
import pcprutils as ut
from statannot import add_stat_annotation
import numpy as np
import cptac.pancan as pc
```

```python
luad = pc.PancanLuad()
```

```python
df = ut.get_prot_trans_df(luad)
df.Tissue = [tissue.title() for tissue in df.Tissue]
df = df.rename(columns = {'Proteomics': 'Protein Abundance', 'Transcriptomics': 'RNA Abundance'})
```

```python
genes = ['NKTR', 'ARID1A', 'ARID1B']
```

```python
diff_prot = pd.read_csv('data/Proteomics_differential_expression_df.csv')
diff_prot = diff_prot[diff_prot.Gene.isin(genes)]
diff_prot = diff_prot[diff_prot.Cancer == 'LUAD']
diff_trans = pd.read_csv('data/Transcriptomics_differential_expression_df.csv')
diff_trans = diff_trans[diff_trans.Gene.isin(genes)]
diff_trans = diff_trans[diff_trans.Cancer == 'LUAD']
delta_corr_df = pd.read_csv('data/delta_correlation_df_with_significance.csv')
```

```python
a = 'Gene: NKTR\n' + 'Cancer: LUAD\n'   + r'$\Delta$' + ' Correlation:' + u'\u2718' + '\n' + 'Diff. Expression: ' + u'\u2714'
b = 'Gene: ARID1A\n' + 'Cancer: LUAD\n' + r'$\Delta$' + ' Correlation:' + u'\u2714' + '\n' + 'Diff. Expression:' + u'\u2714'
c = 'Gene: ARID1B\n' + 'Cancer: LUAD\n'  + r'$\Delta$' + ' Correlation:' + u'\u2714' + '\n' + 'Diff. Expression:' + u'\u2718'


notes = [a,b,c]
```

```python
def axis_percentage(num,percent):
    return percent*num/100
```

```python
sns.set_context(rc={"font.size":16,"axes.titlesize":16,"axes.labelsize":16, 'xtick.labelsize': 12, 'ytick.labelsize': 12}) 
```

```python
fig, ax = plt.subplots(nrows = 3, ncols = 3, figsize=(15,15))
plt.subplots_adjust(hspace = 0.3)
i = 0
for gene, s in zip(genes, notes):
    gene_df = df[df.Gene == gene]
    trans_pval = diff_trans[diff_trans.Gene == gene].FDR
    trans_pval = np.array(trans_pval)
    prot_pval = diff_prot[diff_prot.Gene == gene].FDR
    prot_pval = np.array(prot_pval)
    plt.sca(ax[i][0])
    plt.gca().text(-0.3, 1, s, transform = plt.gca().transAxes, 
                 fontsize = 16, va = 'top', ha = 'right')
    
    #Delta Correlation plot
    reg_normal = sns.regplot(data = gene_df[gene_df.Tissue == 'Normal'], y = 'Protein Abundance', 
                   x = 'RNA Abundance')
    reg_tumor = sns.regplot(data = gene_df[gene_df.Tissue == 'Tumor'], y = 'Protein Abundance', 
                   x = 'RNA Abundance')
    
    #Adding Delta Correlation p_value textbox
    textstr = 'P value = ' + "{:.4f}".format(delta_corr_df[delta_corr_df.Gene == gene][delta_corr_df[delta_corr_df.Gene == gene].Cancer == 'LUAD'].FDR.iloc[0])
    x_max = reg_tumor.axes.get_xlim()[1]
    y_max = reg_tumor.axes.get_ylim()[1]
    if(gene == 'NKTR'):
        text_x_loc = axis_percentage(x_max, 68) 
    if(gene == 'ARID1A'):
        text_x_loc = axis_percentage(x_max, 58.7)
    if(gene == 'ARID1B'):
        text_x_loc = axis_percentage(x_max, 67.4)
    text_y_loc = axis_percentage(y_max, 90)
    reg_tumor.axes.text(text_x_loc,text_y_loc, textstr,fontsize=12)
    
    #Protein differential plot
    plt.sca(ax[i][1])
    g = sns.boxplot(data = gene_df, x = 'Tissue', y = 'Protein Abundance', order = ['Normal', 'Tumor'], fliersize = 0)
    test_results = add_stat_annotation(g, data=df, x='Tissue', y='Protein Abundance', order = ['Normal', 'Tumor'],
                                   box_pairs=[('Normal', 'Tumor')],
                                   pvalues = prot_pval, text_format='star',
                                   loc='outside', perform_stat_test=False, verbose = False)
    
    #RNA differential plot
    plt.sca(ax[i][2])
    g2 = sns.boxplot(data = gene_df, x = 'Tissue', y = 'RNA Abundance', order = ['Normal', 'Tumor'], fliersize = 0)
    test_results = add_stat_annotation(g2, data=df, x='Tissue', y='RNA Abundance', order = ['Normal', 'Tumor'],
                                   box_pairs=[('Normal', 'Tumor')],
                                   pvalues = trans_pval, text_format='star',
                                   loc='outside', perform_stat_test=False, verbose = False)
    i += 1

fig.tight_layout()
plt.savefig('Figures/Figure_2_Correlation_vs_Differential_Expression.png',dpi = 600)
```
