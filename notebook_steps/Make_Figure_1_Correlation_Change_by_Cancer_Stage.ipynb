## Correlation Change by Cancer Stage


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Import dataframe with correlations<br >
Set up dataframe to divide normal and cancer correlations

```python
df = pd.read_csv('data/Cancer_stages_correlations.csv')
df = df[df.Stage != 'Other']
df = df.rename(columns={"Cancer": "Cancer", "Stage" : "Tissue type", "Gene":"Gene", "Spearman_Rho" : "Spearman_Rho"})
```

```python
sns.set_context(rc={"font.size":16,"axes.titlesize":16,"axes.labelsize":16, 'xtick.labelsize': 12, 'ytick.labelsize': 12}) 
```

Plot image

```python
g = sns.FacetGrid(data = df, col = 'Cancer', hue = 'Tissue type', col_wrap = 3, legend_out = True, 
                  palette={"Normal" : "royalblue","Cancer" : "red"})

g.map_dataframe(sns.kdeplot, x = 'Spearman_Rho')
axes = g.axes.flatten()
axes[0].set_title("CCRCC")
axes[1].set_title("Endometrial")
axes[2].set_title("HNSCC")
axes[3].set_title("LSCC")
axes[4].set_title("LUAD")

g.add_legend(bbox_to_anchor=(.75, .25))
g.tight_layout()

plt.savefig('Figures/Figure_1_Correlation_changes_by_cancer_stage', dpi = 600)
plt.show() 
```
