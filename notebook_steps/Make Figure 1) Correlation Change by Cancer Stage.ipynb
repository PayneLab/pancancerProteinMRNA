```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/Cancer_stages_correlations.csv')
df = df[df.Stage != 'Other']
```

```python
sns.set_context(rc={"font.size":16,"axes.titlesize":16,"axes.labelsize":16, 'xtick.labelsize': 12, 'ytick.labelsize': 12}) 
```

```python
g = sns.FacetGrid(data = df, col = 'Cancer', hue = 'Stage', col_wrap = 3, legend_out = True, 
                  palette={"Normal" : "royalblue","Stage I" : "red", "Stage II" : "darkred",
                           "Stage III" : "purple","Stage IV" : "black"})

g.map_dataframe(sns.kdeplot, x = 'PearsonR')
g.add_legend()
g.tight_layout()
plt.savefig('Figures/Figure 1) Correlation changes by cancer stage')
plt.show() 
```
