```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/Cancer_stages_correlations.csv')
df = df[df.Stage != 'Other']
```

```python
sns.set_context(rc={"font.size":16,"axes.titlesize":16,"axes.labelsize":16, 'xtick.labelsize': 12, 'ytick.labelsize': 12}) 
```

```python
g = sns.FacetGrid(data = df, col = 'Cancer', hue = 'Stage', col_wrap = 3, legend_out = True, margin_titles = True,
                  palette={"Normal" : "royalblue","Stage I" : "red", "Stage II" : "darkred",
                           "Stage III" : "purple","Stage IV" : "black"})



g.map_dataframe(sns.kdeplot, x = 'PearsonR')
g.set_titles(col_template = "{col_name}", row_template = "")
g.set_axis_labels('Correlation', 'Density')
g.add_legend(bbox_to_anchor=(.75, .25))
g.tight_layout()
plt.savefig('Figures/Figure 1) Correlation changes by cancer stage')
plt.show() 
```

```python
df = pd.read_csv('data/Cancer_stages_correlations_spearman.csv')
df = df[df.Stage != 'Other']
```

```python
g = sns.FacetGrid(data = df, col = 'Cancer', hue = 'Stage', col_wrap = 3, legend_out = True, margin_titles = True,
                  palette={"Normal" : "royalblue","Stage I" : "red", "Stage II" : "darkred",
                           "Stage III" : "purple","Stage IV" : "black"})

g.map(sns.kdeplot, 'PearsonR')
g.set_titles(col_template = "{col_name}", row_template = "")
g.set_axis_labels('Correlation', 'Density')
g.add_legend(bbox_to_anchor=(.75, .25))
g.tight_layout()
plt.savefig('Figures/Figure 1) Correlation changes by cancer stage')
plt.show() 
```
