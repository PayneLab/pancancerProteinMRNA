```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/Cancer_stages_correlations.csv')
df
```

```python
g = sns.FacetGrid(data = df, col = 'Cancer', hue = 'Stage', col_wrap = 2, legend_out = True)
g.map_dataframe(sns.kdeplot, x = 'PearsonR')
g.add_legend()
plt.savefig('Figures/Figure 1) Correlation changes by cancer stage')
plt.show()
```
