# Overlap in Statistical Significance

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
corr_df = pd.read_csv('data/delta_correlation_df.csv')
```

```python
corr_df['Statistical_Significance'] = corr_df.FDR < 0.05
```

Plot image

```python
colors = ['gray', 'orange']
sns.set_palette(sns.color_palette(colors))
g = sns.FacetGrid(data = corr_df, col= 'Cancer', col_wrap= 3, hue = 'Statistical_Significance', hue_order=[True, False])
g.map_dataframe(sns.histplot, x = 'Delta_Correlation')
g.add_legend()
plt.savefig('Figures/Figure_S4_Significance_Distribution',dpi = 600, bbox_inches='tight')
```
