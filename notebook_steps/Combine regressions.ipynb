```python
import pandas as pd
import glob
```

```python
files = glob.glob('data/*regressions.csv')
files
```

```python
dfs = []
for file in files:
    df = pd.read_csv(file)
    #print(df.columns)
    dfs.append(df)
df = pd.concat(dfs)
df
```

```python
df.to_csv('data/regressions_data.csv', index = False)
```
