```python
import pandas as pd
import scipy
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
permutations = pd.read_csv('full_10k_permutation.csv')
```

```python
pval = pd.read_csv('p_val.csv')
pval = pval[pval.Cancer != 'gbm']
```

```python
cor_diff = pd.read_csv('corr_diff.csv')
cor_diff = cor_diff[cor_diff.Cancer != 'gbm']
```

```python
cor_diff = cor_diff.melt(id_vars = ['Cancer'], var_name = 'gene', value_name= 'cor_diff' )
pval = pval.melt(id_vars = ['Cancer'], var_name = 'gene', value_name= 'transformed_pval' )
permutations = permutations.melt(id_vars = ['Cancer'], var_name = 'gene', value_name= 'permutation_pval' )
```

```python
both_pvals = pd.merge(permutations, pval)
both_pvals = both_pvals.dropna()
both_pvals
```

```python
sns.regplot(data = both_pvals, x = 'permutation_pval', y = 'transformed_pval')
```

```python
perm_corr_diff = pd.merge(permutations, cor_diff)
perm_corr_diff = perm_corr_diff.dropna()
perm_corr_diff
```

```python
sns.regplot(data = perm_corr_diff, y = 'permutation_pval', x = 'cor_diff')
```

```python
perm_corr_diff['cor_diff'] = abs(perm_corr_diff.cor_diff)
```

```python
sns.regplot(data = perm_corr_diff, y = 'permutation_pval', x = 'cor_diff')
```

```python
transform_corr_diff = pd.merge(pval, cor_diff)
transform_corr_diff = transform_corr_diff.dropna()
transform_corr_diff
```

```python
sns.regplot(data = transform_corr_diff, x = 'cor_diff', y = 'transformed_pval')
```

```python
transform_corr_diff['cor_diff'] = abs(transform_corr_diff.cor_diff)
```

```python
sns.regplot(data = transform_corr_diff, x = 'cor_diff', y = 'transformed_pval')
```
