```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
trans_mutation_df = pd.read_csv('data (new)/transmutation_df.csv')
delta_correlation_df = pd.read_csv('data (new)/delta_correlation_df.csv')
```

```python
delta_correlation_df = delta_correlation_df[delta_correlation_df.FDR < 0.05]
delta_correlation_df = delta_correlation_df[['Gene', 'Cancer']]
trans_df = pd.merge(trans_mutation_df, delta_correlation_df, how = 'inner')
trans_df
```

```python
fig, axes = plt.subplots(nrows = 3, ncols = 2, sharex = True, sharey = True,
                        figsize=(12,15))
cancers = list(pd.unique(trans_df.Cancer))
for ax in axes.flatten():
    if len(cancers) == 0:
        break
    cancer = cancers[0]
    cancers.remove(cancer)
    df = trans_df[trans_df.Cancer == cancer]
    plt.subplot(ax)
    sns.scatterplot(data = df, x = 'Delta_Correlation', y= 'FDR', hue = 'Mutated_Gene')
    ax.invert_yaxis()
    ax.set_yscale('log')
    plt.axhline(y = 0.05, color = 'r')
    plt.title(cancer)
```
