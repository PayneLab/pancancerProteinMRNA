```python
import cptac
import cptac.utils as ut
import scipy
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import math
import pandas as pd
```

```python
perm_df = pd.read_csv("full_10k_permutation.csv", index_col=0)
```

```python
aList = []
for i in perm_df:
    count = 0
    for j in perm_df[i]:
        if j <= 0.05/13669:
            count += 1
            if i not in aList and count >= 4:
                aList.append(i)
#     print(perm_df[i])
len(aList)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import gseapy as gp
from gseapy.plot import barplot, dotplot
```

```python
tumor_enr = gp.enrichr(gene_list = aList, description='', gene_sets='MSigDB_Hallmark_2020')
```
